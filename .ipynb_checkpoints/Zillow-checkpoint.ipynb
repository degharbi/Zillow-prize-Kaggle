On remarque du diagram de bar horizontal qu'une bonne parties des variables manquent de donnees

```python
X16_missing = X16_train.isnull().sum(axis=0).reset_index()
X16_missing.columns = ['column_name', 'missing_count']
# X16_missing = X16_missing.ix[missing_df['missing_count']>0]
X16_missing = X16_missing.sort_values(by='missing_count')
```

```python
X16_missing.head
```

```python
X16_missing['missing_rate'] = X16_missing['missing_count']/ X16_train.parcelid.count()
X16_missing = X16_missing.sort_values(by='missing_rate')
X16_missing.head
```

```python
missing_pct=0.9
X16_train.drop(X16_missing[(X16_missing.missing_rate>=missing_pct)].column_name.values.tolist(),
                    axis=1,inplace=True)
```

```python
plt.figure(figsize=(12,12))
sns.set(style="white")

# correlation matrix
corr = X16_train.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3,
            square=True,
            linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig('corr-train-data.eps',bbox_inches='tight')
```
